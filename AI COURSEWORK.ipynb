## 1) The below code gets the data for all the data files included within the data set

```python
import seaborn as snes
import warnings
warnings.filterwarnings("ignore")

f = open("wdbc/wdbc.csv","rt")
wdbc = []
for line in f:
    wdbc.append(line.replace("\n","").replace("\ufeff","").split(","))
f.close()

f = open("wpbc/wpbc.csv","rt")
wpbc = []
for line in f:
    wpbc.append(line.replace("\n","").replace("\ufeff","").split(","))
f.close()

f = open("wdbc/wdbc.csv","rt")
wdbc = []
for line in f:
    wdbc.append(line.replace("\n","").replace("\ufeff","").split(","))
f.close()

f = open("Breast Cancer Wisconsin/BCW.csv","rt")
wdbc = []
for line in f:
    wdbc.append(line.replace("\n","").replace("\ufeff","").split(","))
f.close()
```

## 2) Still trying to get used to using seaborn

```python
%matplotlib notebook
dataN = [[],[]]
dataR = [[],[]]
for record in wpbc:
    if record[1] == "N":
        dataN[0].append(float(record[20]))
        dataN[1].append(float(record[21]))
    else:
        dataR[0].append(float(record[20]))
        dataR[1].append(float(record[21]))

snes.regplot(dataN[0],dataN[1], color = "coral")
snes.regplot(dataR[0],dataR[1], color = "lightseagreen")
#LOVELY WARNING SIGNS
```
