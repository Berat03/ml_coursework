### Messing around with k nearest neighbours, lets start by comparing radius and texture with belign/malignant. Also Benign/Malignant requires label encoding

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
radArray = []
texArray = []

for item in dfWDBC["Radius"]:
    radArray.append([item])

for item in dfWDBC["Texture"]:
    texArray.append([item])


rad = ss.fit_transform(radArray).flatten()
tex = ss.fit_transform(texArray).flatten()
print(rad.flatten())
print(tex.flatten())

le = preprocessing.LabelEncoder()
le.fit_transform(dfWDBC["B/M"])
```

### So now bening/Malignant is represented with 0 and 1

```python
%matplotlib notebook
#Now we have a simple features and label var
features=list(zip(rad,tex))
label = le.fit_transform(dfWDBC["B/M"])

#Now fitting the model to our data 
model = KNeighborsClassifier(n_neighbors=3)
model.fit(features,label)

#All the rest is plotting the data
import numpy as np

x = np.linspace(-3,3,500)
y = np.linspace(-3,3,500)

classOne = []
classTwo = []

for i in x:
    for j in y:
        coor = [i,j]
        if model.predict([coor]) == [1]:
            classOne.append(coor)
        else:
            classTwo.append(coor)

classOneX = []
classOneY = []
for point in classOne:
    classOneX.append(point[0])
    classOneY.append(point[1])
snes.scatterplot(classOneX,classOneY, color = "paleturquoise", size = 0.02, linewidth=0)

classTwoX = []
classTwoY = []
for point in classTwo:
    classTwoX.append(point[0])
    classTwoY.append(point[1])
snes.scatterplot(classTwoX,classTwoY, color = "lightsalmon",size = 0.02, linewidth=0)
```

# Model 1, K-Neighbours to classify B/M using the features Radius and Texture

## <span style="color: red;">STEP 1) Split the dataSet up into test and training sets</span> 

```python
#Getting a dataFrame with only the info we care about
model_data = dfWDBC[["Radius","Texture","B/M"]]

from sklearn.model_selection import train_test_split
features = model_data[["Radius","Texture"]]
label = model_data["B/M"]

x_train,x_test,y_train,y_test = train_test_split(features,label,test_size= 0.33)
```

## <span style="color: red;">Step 2) Fit the scaler to the training data and scale all info off that</span> 

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()

#converting the array structure of each feature
rad_train = []
for item in x_train["Radius"]:
    rad_train.append([item])
rad_test = []
for item in x_test["Radius"]:
    rad_test.append([item])
    
ss.fit(rad_train)

rad_train_fitted = ss.transform(rad_train).flatten()
rad_test_fitted = ss.transform(rad_test).flatten()
#Radius is now standardised

tex_train = []
for item in x_train["Texture"]:
    tex_train.append([item])
tex_test = []
for item in x_test["Texture"]:
    tex_test.append([item])
    
ss.fit(tex_train)

tex_train_fitted = ss.transform(tex_train).flatten()
tex_test_fitted = ss.transform(tex_test).flatten()

```

## <span style="color: red;">Step 3) Encode the label and fit the classifier model</span> 

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
label_train = le.fit_transform(y_train)
label_test = le.fit_transform(y_test)

features_train = list(zip(rad_train_fitted,tex_train_fitted))

from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier(n_neighbors=40)
model.fit(features_train,label_train)
```

## <span style="color: red;">Step 4) Plot the predictions</span> 

```python
import numpy as np
%matplotlib notebook

x = np.linspace(-5,5,600)
y = np.linspace(-5,5,600)

classOne = []
classTwo = []

for i in x:
    for j in y:
        coor = [i,j]
        if model.predict([coor]) == [1]:
            classOne.append(coor)
        else:
            classTwo.append(coor)

classOneX = []
classOneY = []
for point in classOne:
    classOneX.append(point[0])
    classOneY.append(point[1])
snes.scatterplot(classOneX,classOneY, color = "paleturquoise", size = 0.001, linewidth=0)

classTwoX = []
classTwoY = []
for point in classTwo:
    classTwoX.append(point[0])
    classTwoY.append(point[1])
snes.scatterplot(classTwoX,classTwoY, color = "lightsalmon",size = 0.001, linewidth=0)

#Plotting the test data above the model we trained
#snes.scatterplot(rad_test_fitted,tex_test_fitted)
d = {"B/M":label_test, "Radius": rad_test_fitted, "Texture":tex_test_fitted}
new_df = pd.DataFrame(d)
dfM_fitted = new_df.loc[new_df["B/M"] == 1]
dfB_fitted = new_df.loc[new_df["B/M"] == 0]
snes.scatterplot(data = dfM_fitted, x = "Radius", y = "Texture", color = "blue")
snes.scatterplot(data = dfB_fitted, x = "Radius", y = "Texture", color = "red")


#CAN TAKE A HOT SECOND TO RUN!!!
```

## <span style="color: red;">Step 5) Test the hyper-parameters</span> 

```python
features_test = list(zip(rad_test_fitted,tex_test_fitted))

label_pred = model.predict(features_test)

from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(label_test, label_pred))
```

## <span style="color: red;">Step 6) Test the hyper-parameters for different numbers of k neighbours, and then for fun plot the accuracy across k</span> 

```python
%matplotlib notebook
from scipy.ndimage.filters import gaussian_filter1d
accuracy = []
kArr = []
n=100
for k in range(1,n):
    model = KNeighborsClassifier(n_neighbors=k)
    model.fit(features_train,label_train)
    label_pred = model.predict(features_test)
    kArr.append(k)
    accuracy.append(metrics.accuracy_score(label_test, label_pred))
    print("Accuracy:",metrics.accuracy_score(label_test, label_pred))
accuracySmoothed = gaussian_filter1d(accuracy, sigma=2)
snes.pointplot(kArr,accuracySmoothed, scale=0.5, color= "green")
print(kArr[accuracy.index(max(accuracy))])
```
