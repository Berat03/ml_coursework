```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

Area and Perimeter are scalars of Radius. We will remove these features to eliminate covariance between our features.

https://datascience.stackexchange.com/questions/87648/elimination-of-features-based-on-high-covariance-without-affecting-performance

Features that have normal distribution:
- Smoothness
- Symmetry

Features that I think are lognormally distributed:
- Texture
- Radius
- FractalDimension
- Concavity
- ConcavePoints
- Compactness

```python
df = pd.read_csv('./wdbc.csv')
features = df[['Radius', 'Texture', 'Smoothness', 'Compactness', 'Concavity', 'ConcavePoints', 'Symmetry', 'FractalDimension']]

fig, ax = plt.subplots(4, 2)
sns.histplot(df['Radius'], ax=ax[0, 0])
sns.histplot(df['Texture'], ax=ax[0, 1])
sns.histplot(df['Smoothness'], ax=ax[1, 0])
sns.histplot(df['Compactness'], ax=ax[1, 1])
sns.histplot(df['Concavity'], ax=ax[2, 0])
sns.histplot(df['ConcavePoints'], ax=ax[2, 1])
sns.histplot(df['Symmetry'], ax=ax[3, 0])
sns.histplot(df['FractalDimension'], ax=ax[3, 1])
```

Of the following features, which have a lognormal distribution?
- Radius - YES
- Texture - YES
- Smoothness - NO (normally distributed)
- Compactness - YES
- Concavity - NO
- ConcavePoints - NO
- FractalDimension - YES
- Symmetry - NO (normally distributed)
